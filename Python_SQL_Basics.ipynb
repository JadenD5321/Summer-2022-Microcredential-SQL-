```python
import pandas as pd
import pandasql as ps
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/JadenD5321/Machine-Learning-Data/main/Social_Network_Ads.csv')
```

```python
df.info()
```

```python
# Query in SQL means to request data for a specific information. In this case, we tried to find the average salary specifically for 20 year old male and females. we will then analyze and see if there are unsual patterns throughout. 



df.query('Age == 20')
```

```python
df.query('Age > 20')
```

```python
df.query('Age < 20')
```

```python
df.query('Gender == "Male"')
```

```python
# Using pandasql

query = "SELECT * FROM df"
ps.sqldf(query)
```

```python
query = "SELECT * FROM df WHERE Gender == 'Male' AND Age == 20"
ps.sqldf(query)
```

```python
query = "SELECT * FROM df WHERE Gender == 'Male' AND EstimatedSalary >= 150000"
ps.sqldf(query)
```

```python
query = "SELECT * FROM df WHERE EstimatedSalary >= 120000"
ps.sqldf(query)
```
